```python
import bamboolib as bam
bam.enable()
```

```python
import pandas as pd
df = pd.read_csv("data/titanic.csv")
```

```python
df
# bamboolib code export - don't write below - rerun cell to execute code
df['Survived'] = df['Survived'].astype(bool, errors='raise')
df['Name'] = df['Name'].astype(str, errors='raise')
split_df = df["Name"].str.split(',', expand=True)
split_df.columns = [f"Name_{id_}" for id_ in range(len(split_df.columns))]
df = pd.merge(df, split_df, how="left", left_index=True, right_index=True)
df = df.rename(columns={'Name_0': 'LastName'})
df = df.rename(columns={'Name_1': 'FirstName'})
df = df.loc[~(df['Age'].isna())]
df.index = pd.RangeIndex(len(df))
```
